```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm

# Create a small custom dataset with 10 data points
X = np.array([
    [1, 2],
    [2, 3],
    [3, 3],
    [2, 1],
    [3, 2],
    [6,6],
    [7,1],
    [5,4],
    [7,8],
    [7, 8],
    [8, 9],
    [9, 8],
    [8, 7],
    [9, 9]
])
y = np.array([0, 0, 0, 0, 0,1,0,1,1 ,1, 1, 1, 1, 1])

# Define different values of C
C_values = [0.1, 1, 10, 100]

plt.figure(figsize=(12, 10))
for i, C in enumerate(C_values):
    # Create an instance of SVM and fit the data.
    clf = svm.SVC(kernel='linear', C=C)
    clf.fit(X, y)

    # Plot the data points
    plt.subplot(2, 2, i + 1)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o')


    # Create grid to evaluate model
    xx = np.linspace(0, 10)
    yy = -(clf.coef_[0][0] * xx + clf.intercept_[0]) / clf.coef_[0][1]

    # Plot hyperplane
    plt.plot(xx, yy, 'k--')

    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title(f'SVM with linear kernel and C={C}')

plt.tight_layout()
plt.show()
```
